# 기술 통계량
- 데이터를 기술하거나 설명하는 숫자
- 복잡한 데이터를 간단하고 명확하게 요약하여 데이터의 특징을 파악하는 데 사용
- 흔히 사용하는 평균, 중앙값, 표준편차 등이 모두 기술 통계량에 속함

## 기술 통계량의 필요성
- 데이터 이해
    - 방대한 양의 데이터를 한눈에 파악하기 어렵기 때문에, 기술 통계량을 통해 데이터의 중심 경향, 분산 정도 등을 간단하게 요약하여 데이터의 전체적인 모습을 파악
- 데이터 비교
    - 서로 다른 데이터 집합 간에 기술 통계량을 비교하여 차이점을 찾아낼 수 있음
- 데이터 분석의 기초
    - 데이터 전처리, 추론을 위한 새로운 특성 생성 등 더 복잡한 분석을 하기 위한 기초 자료로 활용

## 대표적인 기술 통계량
- 중심 경향
    - 데이터의 중심 위치를 나타내는 값
    - 평균, 중앙값, 최빈값
- 분산 정도
    - 범위, 분산, 표준편차
- 분포의 모양
    - 왜도, 첨도

## 기술 통계의 한계점
- 데이터의 분포
    - 평균의 함정
    > 평균은 데이터의 중심값을 나타내지만, 극단적인 값에 민감하게 반응하여 전체 데이터의 분포를 정확하게 반영하지 못할 수 있음
    - 분산과 표준편차
    > 데이터가 얼마나 퍼져 있는지를 나타내지만, 분포의 형태에 대한 정보는 제공하지 않음
- 인과관계
    - 상관관계와 인과관계
    > 두 변수 간의 상관관계가 높다고 해서 한 변수가 다른 변수의 원인이 되는 것으로 해석하면 안됨
- 데이터의 질
    - 데이터 수집 과정
    > 데이터를 수집하는 과정에서 오류나 편향이 발생할 수 있음
    - 데이터의 정확성
    > 데이터 입력 오류, 측정 오류 등으로 인해 데이터 자체가 부정확할 수 있음
- 복잡한 현상
    - 단순화
    > 기술 통계는 복잡한 현상을 단순화된 숫자나 그래프로 표현하기 때문에, 현상의 모든 측면을 포착하지 못함
    - 상호작용 효과
    > 여러 변수 간의 상호작용 효과를 파악하기 어려움

# 데이터의 종류와 특징
- 데이터는 크게 수치형 데이터와 범주형 데이터로 나눌 수 있음
- 각각의 데이터는 가지는 특징과 분석 방법이 다르기 때문에, 데이터 분석을 시작하기 전에 이 두 가지를 명확하게 구분하는 것이 중요


    

## 수치형 데이터 (Numerical Data)
- 수치형 데이터는 숫자로 표현되는 데이터
- 수치형 데이터는 다시 연속형 데이터와 이산형 데이터로 나눌 수 있음
- 연속형
    - 특정 범위 내에서 어떤 값이든 취할 수 있는 데이터
    - 연속형 데이터는 소수점 이하 자리까지 측정할 수 있으며, 무한히 많은 값을 가질 수 있음
    - 키, 몸무게, 온도, 시간, 풍속 등
- 이산형
    - 정수 값만을 가지는 데이터
    - 이산형 데이터는 셀 수 있는 값만을 가지며, 연속적인 값을 가질 수 없음
    - 사건의 발생 빈도 같은 횟수,방개수, 자녀 수, 자동차 수, 판매량 등

## 범주형 데이터 (Categorical Data)
- 가능한 범주의 값만 취할수 있는 데이터
- 범주형 데이터는 다시 명목형 데이터와 순서형 데이터로 나눌 수 있음
- 명목형
    - 범주 간에 순서가 없는 데이터
    - 성별(남/여), 혈액형(A/B/O/AB), 국적, 전공(경영학/인문학/교육학), 장르(공포/스릴러/드라마) 등
- 순서형
    - 범주 간에 순서가 있는 데이터
    - 학점(A, B, C, D), 영화평점(1, 2, 3, 4, 5), 만족도(매우 만족, 만족, 보통, 불만족), 직급(대리, 과장, 차장, 부장)

    - 평균 하나만으로는 데이터의 분포나 특징을 완벽하게 파악하기 어려움

## 중앙값(median)
- 어떤 데이터 집합을 크기 순서대로 나열했을 때 가장 중앙에 위치하는 값
- 데이터를 반으로 나누었을 때 위아래 그룹의 경계선에 있는 값
- 왜 중앙값을 사용할까?
    - 평균은 극단적으로 큰 값이나 작은 값에 영향을 받기 쉽지만, 중앙값은 데이터의 중심 경향을 나타내는 데 있어 극단값의 영향을 덜 받음
    - 평균과 같이 사용하면 데이터가 어떻게 분포되어 있는지 파악하는 데 도움이 됨

## 분산 (variance)
- 분산은 통계학에서 데이터가 평균으로부터 얼마나 흩어져 있는지를 나타내는 값
- 데이터의 흩어진 정도를 수치로 나타낸 것
- 각 데이터 값에서 평균을 빼서 구한 편차들을 제곱하여 평균
- 왜 분산을 사용할까?
    - 분산은 데이터가 얼마나 넓게 퍼져 있는지를 알려주어 데이터의 특징을 더 정확하게 파악
    - 서로 다른 데이터 집합의 흩어진 정도를 비교

```python
df['연간소득'].var()
```

```python
df.loc[mask, '연간소득'].var(), df.loc[~mask, '연간소득'].var()
```

## 왜도(Skewness)
- 데이터의 분포의 비대칭도를 나타내는 통계량
- 데이터가 얼마나 대칭에서 벗어나 있는지를 나타내는 척도
- 양의 왜도
    - 데이터가 오른쪽으로 긴 꼬리를 가지고 있을 때 나타남
    - 작은 값은 많고 큰 값은 적은 분포를 의미
    - 평균이 중앙값보다 크게 나타
- 음의 왜도
    - 데이터가 왼쪽으로 긴 꼬리를 가지고 있을 때 나타남
    - 큰 값은 많고 작은 값은 적은 분포를 의미
    - 평균이 중앙값보다 작게 나타남
- 왜도가 0
    - 데이터가 대칭 분포일 때 나타남
    - 평균, 중앙값이 거의 일치

## 첨도(Kurtosis)
- 첨도는 데이터 분포의 뾰족한 정도를 나타내는 통계량
- 데이터가 얼마나 중심에 집중되어 있고, 꼬리가 얼마나 두꺼운지를 나타내는 척도
- 양의 첨도
    - 분포의 꼬리가 두껍고, 중심에 데이터가 많이 몰려 있음을 의미
- 음의 첨도
    - 분포의 꼬리가 얇고, 중심에서 데이터가 덜 몰려 있음을 의미
- 첨도가 0
    - 표본 분포가 정규분포와 같은 형태로, 꼬리의 두께가 정규분포와 비슷함을 의미

## 상관관계
- 두 수치형 변수의 관계를 나타내는 수치
- 두 변수가 얼마나 밀접하게 연관되어 있는지를 나타내는 척도
- 두 수치형 변수의 변화가 서로 관계가 있을 때 상관관계가 있다고 함
- 양의 상관관계 (+)
    - 한 변수가 큰 값을 가질 때 다른 변수도 큰 값을 갖는가?
    - 한 변수의 값이 증가하면 다른 변수의 값도 증가하는 경우
- 음의 상관관계 (-)
    - 아니면 오히려 작은 값을 갖는가?
    - 한 변수의 값이 증가하면 다른 변수의 값은 감소하는 경우
- 무상관 (0)
    - 아니면 아무런 관계가 없는가?
    - 관계가 없음을 의미

### 피어슨 상관계수 (Pearson correlation coefficient)
- 연속형 변수 간의 선형 관계를 측정하는 가장 일반적인 상관계수
- 두 변수가 얼마나 밀접하게 연관되어 있는지, 그리고 그 관계가 양의 방향인지 음의 방향인지를 수치로 표현
- 언제나 -1과 1 사이의 값을 가진다.
- Pearson 상관계수의 특징
    - 직선관계의 정도를 나타낸다.
    - cor>0: 같은 방향으로 움직이는 경향
    - cor<0: 반대 방향으로 움직이는 경향
    - +-1 에 가까울 수록 (즉, 절대값이 클수록) 강한 상관관계
        - cor=1: 완벽한 직선관계를 의미
        - cor=0에 가까울 수록 관계없음

## 켄달의 타우와 스피어만 상관계수
- 켄달의 타우와 스피어만 상관계수는 모두 두 변수 간의 순위 상관관계를 측정
- -1부터 1 사이의 값
- 부호
    - +1: 완벽한 양의 순위 상관관계 (한 변수의 순위가 높을수록 다른 변수의 순위도 높음)
    - -1: 완벽한 음의 순위 상관관계 (한 변수의 순위가 높을수록 다른 변수의 순위는 낮음)
    - 0: 순위 간에 관계가 없음

- 켄달의 타우 (Kendall's tau)
    - 두 변수의 순위쌍을 비교하여 일치하는 쌍과 불일치하는 쌍의 수를 세어 상관관계를 계산
    - 순위가 얼마나 일치하는지를 직접적으로 비교
- 스피어만 상관계수 (Spearman rank correlation coefficient)
    - 두 변수의 순위 차이의 제곱합을 계산하여 상관관계를 구함
    - 순위 차이의 크기를 기반으로 상관관계를 측정



```python
df['근로기간'].corr(df['대출금액'], method = 'spearman')
```

```python
df['근로기간'].nunique()
```

```python
# 스피어만 상관계수
df['신용거래기간'].corr(df['근로기간'], method = 'spearman')
```
