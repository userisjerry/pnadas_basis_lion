* 단일 컬럼에 대해 다양하게 집계하면서 집계한 데이터에 컬럼명들을 정의하면서 집계

## merge
- 두 개의 데이터프레임을 공통된 열(컬럼)을 기준으로 병합하는 데 사용
- how 파라미터
    - inner(기본값)
        - 기준이 되는 컬럼의 데이터로 merge 하는데 같은 값이 아닌 행들은 사라져서 반환
        - 두 DataFrame에 모두 존재하는 데이터만 병합
    - left
        - 기준이 되는 컬럼의 데이터로 merge 하는데 왼쪽 데이터 프레임의 같은 값이 아닌 행들이 사라지지 않고 반환
        - 왼쪽 DataFrame의 모든 데이터를 포함하고, 오른쪽 DataFrame에 없는 데이터는 NaN으로 채움

    ```python
    pd.merge(left, right, on='기준 컬럼', how='left')
    ```

# apply 메소드 사용해보기
- Pandas의 apply 메서드는 데이터프레임의 특정 축(행 또는 열)에 사용자 정의 함수를 적용하여 새로운 데이터를 생성하거나 기존 데이터를 변환하는 데 사용
- Series, DataFrame에 단순한 집계가 아닌 구체적인 로직을 적용하고 싶을 때 활용
- 단순 집계 뿐이 아닌 각 데이터에 조건검사등 복잡한 처리가 가능
- apply에 전달한 콜백함수에 데이터를 인수로 전달한다.


````python
cols = [ col for col in df.columns if col.startswith("캠페인") ]
cols
````

````python
# 고객_교육수준 별로 이전 캠페인들의 수락의 합을 보고 싶다면?
def do_apply(x):
    cols = [ col for col in df.columns if col.startswith("캠페인") ]
    return x[cols].sum().sum()

df.groupby("고객_교육수준").apply(do_apply)
````

* 콜백 함수에 추가 파라미터 정의해보기
