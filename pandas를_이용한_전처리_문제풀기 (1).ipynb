```
고유한 고객번호로만 이루어진 데이터프레임을 만들어서 features 변수에 넣으시오.
컬럼명은 ID로 하시오.
컬럼이 ID만 있는 데이터프레임을 만들라는 의미
중복 고객번호가 있으면 안된다.
```

````python
# 고유한 고객번호로만 - df['ID'].unique()
# 데이터 프레임으로 만들어 - pd.DataFrqme(df['ID'].unique(), columns=['ID'])
````

````python
features = pd.DataFrame(df['ID'].unique(), columns=['ID'])
features
type(features)
````

```
groupby 메소드를 이용하여 고객별로 구매가격의 평균을 구하고 데이터 프레임 형식으로 정돈 하고 구매가격의 컬럼명을 변경하여
features 변수의 데이터프레임과 left merge 하시오.
```

````python
# feature를 만들어서 left

# 고객별 - df.groupby('ID')
# 구매가격의 평균 - df.groupby('ID')['구매가격'].mean()
# left merge 즉, column으로 생성 -> df.groupby('ID')['구매가격'].mean().reset_index()
# column 명을 바꿔줘 - df.groupby('ID')['구매가격'].mean().reset_index().rename(columns={'구매가격' : '구매가격_ㅡmean'})

tmp = df.groupby('ID')['구매가격'].mean().reset_index().rename(columns={'구매가격' : '구매가격_ㅡmean'})
features = pd.merge(features, tmp, how='left', on='ID')
features
## 샘플수가 절대 늘어나면 안돼 -> 중복된것이 생긴것 -> row 항상 확인
## 늘어나면 groupby에서 잘못한것 ;(예시 - df.groupby(['ID','대분류'])['구매가격'].)
````

```
groupby 메소드를 이용하여 고객별로 구매가격의 최대값과 대분류의 고유값의 개수(다양성)를
데이데이터 프레임 형식으로 정돈 하고 구매가격과 대분류의 컬럼명을 변경하여
features 변수의 데이터프레임과 left merge 하시오
```

````python
agg_dict = {
    '구매가격' : 'max',  ## 구매가격의 최대값
    '대분류' : 'nunique' ##대분류의 고유값의 개수(다양성)
}

tmp = df.groupby('ID').agg(agg_dict).reset_index()
tmp.columns = ['ID', '구매가격_max', '대분류_nunique' ]

features = pd.merge(features, tmp, how='left', on='ID')
features
````

````python
tmp = df.groupby('ID').agg(agg_dict)
tmp.columns = [ '구매가격_max', '대분류_nunique' ]

features = pd.merge(features, tmp.reset_index(), how='left', on='ID')
features

## ? 난 왜 column이 6개야
````

```
groupby 메소드를 이용하여 고객별로 구매가격의 최소값과, 총 구매액
그리고 지점코드의 고유값 개수와 최빈값을 구하시오
구하는 과정에서 집계 컬럼에 대한 컬럼명도 정의하시오.
데이터프레임형식으로 정돈하고 features 변수의 데이터프레임과 left merge 하시오.
```

````python
agg_dict = {
    '구매가격' : [
        ('구매가격_min', lambda x : x[x>0].min()),
        ('총구매액', 'sum')
    ],
    '지점코드' : [
        #('컬럼명', '집계문자열 또는 콜백함수')
          ('지점코드_nunique', 'nunique'),
        ('지점코드_mode', lambda x : x.mode()[0])
    ]

}
tmp = df.groupby('ID').agg(agg_dict)
tmp
````

````python
tmp.columns = tmp.columns.droplevel()
tmp.columns
````

````python
features = pd.merge(features, tmp.reset_index(), how='left', on='ID')
features
````
